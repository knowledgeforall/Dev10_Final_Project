```python
import pandas as pd
```

```python
districts = pd.read_csv('FederallyFundedDistricts.csv', encoding= 'unicode_escape')
districts.columns
```

```python
# Isolate needed columns

nydistricts = districts[['School District Name','General Fund Total Expenditures & Transfers (Federal)']]
nydistricts.info()
```

```python
# Update columns to match District Names

nydistricts['School District Name'] = nydistricts['School District Name'].str.upper()
nydistricts.rename(columns = {'School District Name' : 'DistrictName'}, inplace = True)
nydistricts.rename(columns = {'General Fund Total Expenditures & Transfers (Federal)' : 'FederallyFunded'}, inplace = True)
nydistricts = nydistricts.replace(',','', regex = True)
nydistricts.head()
```

```python
# Update columns to match District Names

districtinfo = pd.read_csv('PreliminaryDistrictDataNYCDistrictsIncome4.csv')
districtinfo = districtinfo.replace('CSD','', regex = True)
districtinfo = districtinfo.replace('UFSD','', regex = True)
districtinfo = districtinfo.replace('CITY SCHOOL DISTRICT','', regex = True)
districtinfo = districtinfo.replace(' ','', regex = True)
districtinfo.head()
```

```python
#Merge dataframes

mergeddf = pd.merge(districtinfo, nydistricts, on = 'DistrictName', how = 'left')
mergeddf.FederallyFunded.unique()
```

```python
mergeddf.head(24)
```

```python
#Update row values to reflect Yes or No

def federalfund(num):
    if num == 0:
        return "No"
    elif num > 0:
        return "Yes"
    else:
        return num
    
mergeddf['FederallyFunded'] = mergeddf['FederallyFunded'].replace(',','', regex = True).apply(lambda x: federalfund(float(x)))
```
