```python
import pandas as pd
```

Import Budget Information for the 32 districts in NYC

```python
nycdistricts = pd.read_csv('../Datasets/NYCSchoolDistricts.csv')
nycdistricts.columns
```

```python
#Isolate two columns needed
nycdistrictsinfo = nycdistricts[['District','Total Budget Allocation']]
nycdistrictsinfo.info()
```

```python
nycdistrictsinfo.head()
```

```python
#Add needed information to empty list

totalbudget = []

for i in nycdistrictsinfo['Total Budget Allocation']:
    totalbudget.append(i)
```

```python
#Clean up data

totalbudget = [x.replace('$', '') for x in totalbudget]
totalbudget = [x.replace(' ', '') for x in totalbudget]
totalbudget = [x.replace(',', '') for x in totalbudget]

nycdistrictsinfo['Total Budget Allocation'] = totalbudget
nycdistrictsinfo.head()
```

```python
#Change column to numeric

nycdistrictsinfo['Total Budget Allocation'] = pd.to_numeric(nycdistrictsinfo['Total Budget Allocation'])
```

```python
#Group by district to get total funding per district

districtbudget = nycdistrictsinfo.groupby('District').sum()
districtbudgetlist = []
for i in districtbudget['Total Budget Allocation']:
    districtbudgetlist.append(i)
print(districtbudgetlist)
```

```python
#Read in districtinfo csv

districtinfo = pd.read_csv('../Datasets/PreliminaryDistrictDataWITHMEDIANINCOME2.csv')
districtinfo.head()
```

```python
#Find what index the NYC Districts are located 

districtinfo[districtinfo['DistrictName'].str.contains('NYC')]
```

```python
#Get the enrollment values in a list

NYCTotalEnrolled = []
for i in districtinfo['TotalEnrolled'].loc[441:473]:
    NYCTotalEnrolled.append(i)
print(NYCTotalEnrolled)
```

```python
#Divide both lists to get funding per student per district

districtfundingperchild = [i / j for i, j in zip(districtbudgetlist, NYCTotalEnrolled)]
districtfundingperchild = [round(num,2) for num in districtfundingperchild]
print(districtfundingperchild)
```

```python
#Update column values 

one = 0

for i in range (441,474):

    districtinfo.at[i, 'FundingPerStudent'] = districtfundingperchild[one]

    one = one + 1
```

```python
#Confirm column was updated
districtinfo[districtinfo['DistrictName'].str.contains('NYC')]
```
