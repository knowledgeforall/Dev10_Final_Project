```
from pyspark.sql.types import DecimalType, StringType, FloatType

district_df = districtdata \
  .withColumn("DistrictName", districtdata["DistrictName"].cast(StringType())) \
  .withColumn("GraduationPercentageRate", districtdata["GraduationPercentageRate"].cast(FloatType())) \
  .withColumn("DropoutPercentageRate", districtdata["DropoutPercentageRate"].cast(FloatType())) \
  .withColumn("FundingPerStudent", districtdata["FundingPerStudent"].cast(FloatType())) \
  .withColumn("TotalEnrolled", districtdata["TotalEnrolled"].cast(FloatType())) \
  .withColumn("MalesEnrolled", districtdata["MalesEnrolled"].cast(FloatType())) \
  .withColumn("FemalesEnrolled", districtdata["FemalesEnrolled"].cast(FloatType())) \
  .withColumn("Percentage of Minority Students", districtdata["Percentage of Minority Students"].cast(FloatType()))
```

```
display(district_df)
```

```
from pyspark.sql.functions import col
ZipCodeDistrict = zipcodedistrict.withColumn("DistrictName",col("DistrictName").cast(StringType()))   .withColumn("ZipCode",col("ZipCode").cast(FloatType()))
ZipCodeDistrict.printSchema()
```

```
ZipCodeIncome = zipcodemedianincome.withColumn("ZipCode",col("ZipCode").cast(FloatType()))   .withColumn("HouseholdIncome",col("HouseholdIncome").cast(FloatType()))
ZipCodeIncome.printSchema()
```
