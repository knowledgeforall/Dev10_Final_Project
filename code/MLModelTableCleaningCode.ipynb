```python
import pandas as pd
import numpy as np
# from sklearn.impute import SimpleImputer
```

```python
# Storing the file path in 'loc' variable
loc = (r"C:\Users\Minet\Documents\Repositories\Dev10_Final_Project\Datasets\districtdatafinal.csv")
# Importing the file from the file path contained in 'loc' into 'df' 
df = pd.read_csv(loc)
# Showing the first 5 rows of 'df'
df.head()
```

```python
# create new dataframe
df = df[['GraduationPercentageRate','HouseholdIncome','FundingPerStudent', 'DropoutPercentageRate', 'Percentage of Minority Students', 'DistrictName']]
df.head()
```

```python
# remove dashes and zero values
df = df[df.DropoutPercentageRate != '—']
df = df[df.GraduationPercentageRate != '—']
df = df[df.GraduationPercentageRate != '0']
```

```python
print(df.isnull().sum())
```

```python
# drop rows with empty values
df = df.dropna()
df.head(13)
```

```python
print(df.isnull().sum())
```

```python
# write the dataframe to a csv file
df.to_csv(r"C:\Users\Minet\Documents\Repositories\Dev10_Final_Project\Datasets\AltMLModelDatasetPowerBI.csv")
```
