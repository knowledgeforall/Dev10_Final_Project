```python
import pandas as pd
```

```python
districts = pd.read_csv('../Datasets/FederallyFundedDistricts.csv', encoding= 'unicode_escape')
districts.columns
```

```python
districts
```

```python
# Isolate needed columns

nydistricts = districts[['School District Name','General Fund Total Expenditures & Transfers (Federal)']]
nydistricts.info()
```

```python
# Update columns to match District Names

nydistricts['School District Name'] = nydistricts['School District Name'].str.upper()
nydistricts.rename(columns = {'School District Name' : 'DistrictName'}, inplace = True)
nydistricts.rename(columns = {'General Fund Total Expenditures & Transfers (Federal)' : 'FederallyFunded'}, inplace = True)
nydistricts = nydistricts.replace(',','', regex = True)
nydistricts.head()
```

```python
# reading in web scraped dataset
districtinfo = pd.read_csv('../Datasets/PreliminaryDistrictDataNYCDistrictsFunding3.csv')
```

```python
# updating district names in districtinfo and saving for later
# this is data cleaning to match district names with the PowerBI map json naming conventions
districtinfo = districtinfo.replace('CITY SCHOOL DISTRICT','CITY SD', regex = True)
districtinfo = districtinfo.replace('COMMON SCHOOL DISTRICT','COMN SD', regex = True)
districtinfo = districtinfo.replace('CENTRAL HS DISTRICT','CENTRAL SCHOOL DISTRICT',regex = True)
districtname = districtinfo['DistrictName']
```

```python
# Update columns to match District Names for merging
districtinfo = districtinfo.replace('CSD','', regex = True)
districtinfo = districtinfo.replace('UFSD','', regex = True)
districtinfo = districtinfo.replace('CITY SD','', regex = True)
districtinfo = districtinfo.replace('COMN SD','', regex = True)
districtinfo = districtinfo.replace('CENTRAL SCHOOL DISTRICT','',regex = True)
districtinfo = districtinfo.replace(' ','', regex = True)
districtinfo.head()
```

```python
#Merge dataframes

mergeddf = pd.merge(districtinfo, nydistricts, on = 'DistrictName', how = 'left')
mergeddf.FederallyFunded.unique()
```

```python
# reapply previous naming convention
mergeddf['DistrictName']=districtname
mergeddf.head(24)
```

```python
#Update row values to reflect Yes or No

def federalfund(num):
    if num == 0:
        return "No"
    elif num > 0:
        return "Yes"
    else:
        return num
    
mergeddf['FederallyFunded'] = mergeddf['FederallyFunded'].replace(',','', regex = True).apply(lambda x: federalfund(float(x)))
```

```python
mergeddf.head(24)
```

```python
# Make sure NYC districts have correct values

mergeddf[mergeddf['DistrictName'].str.contains('NYC')]
```

```python
#Update NYC Districts to be federally funded

one = 0

for i in range (441,473):

    mergeddf.at[i, 'FederallyFunded'] = 'Yes'

    one = one + 1
```

```python
# clean GraduationPercentageRate and DropoutPercentageRate by converting to float
# to_numeric(errors="coerce") changes dashes into NaN values
mergeddf['GraduationPercentageRate'] = pd.to_numeric(mergeddf['GraduationPercentageRate'], errors='coerce')
mergeddf['DropoutPercentageRate'] = pd.to_numeric(mergeddf['DropoutPercentageRate'], errors='coerce')
mergeddf
```
