```python
import pandas as pd
import numpy as np

df = pd.read_csv("grf21_lea_zcta5ce20.csv")
display(df)
```

```python
school_district_zip_code = df[['NAME_LEA21','ZCTA5CE20']]
display(school_district_zip_code)
```

```python
school_district_zip_code = school_district_zip_code.rename(columns={'NAME_LEA21':'School District Name'})
school_district_zip_code = school_district_zip_code.rename(columns={'ZCTA5CE20':'ZipCode'})
display(school_district_zip_code)
```

```python
school_district_zip_code = school_district_zip_code[school_district_zip_code['ZipCode'].between(10001, 14905)]
display(school_district_zip_code)
```

```python
school_district_zip_code.drop_duplicates(subset=['ZipCode','School District Name'], keep=False)
```

```python
school_district_zip_code = school_district_zip_code.set_index('ZipCode')
display(school_district_zip_code)
```

```python
school_district_zip_code.to_csv('school_district_zip_code.csv')
```
