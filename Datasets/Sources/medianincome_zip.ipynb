```python
# import necessary modules
import pandas as pd
import numpy as np
```

```python
# read in the raw csv dataset
df = pd.read_csv("ACSST5Y2020.S1902-2022-08-03T154159.csv")
df.head(3)
```

```python
# eliminate all rows but desired one
df = df.loc[[1]]
df.head(3)
```

```python
# DF['id2'] = DF.index
# DF.set_index('id',inplace=True)
```

```python
# switch columns and rows
df = df.transpose()
df.head(3)
```

```python
# #Create appropriate labels for header

# #grab the first row for the header
new_header = df.iloc[0]

# #take the data other than the header row
df = df[1:]

# #set the header row as the df header
df.columns = new_header
df.head(3)
```

```python
# Change index name to 'ZipCodes'
df.index.names = ['ZipCodes']
df.head(3)
```

```python
# change column name to 'MedianHouseholdIncome'
df.columns = ['MedianHouseholdIncome']
df.head(3)
```

```python
# change index to column
df.reset_index(inplace=True)
df.head(3)
```

```python
# strip unwanted left text from 'ZipCodes' column
df['ZipCodes'] = df['ZipCodes'].map(lambda x: x.lstrip('+-').lstrip('ZCTA5 '))
df.head(3)
```

```python
# strip unwanted right text from 'ZipCodes' column
df['ZipCodes'] = df['ZipCodes'].map(lambda x: x.lstrip('+-').rstrip('!!Mean income (dollars)!!Estimate'))
df.head(3)
```
